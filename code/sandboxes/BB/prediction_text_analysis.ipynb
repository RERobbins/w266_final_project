### Imports/Setup

```python
from os import listdir
from os.path import isfile, join

import csv
import json
import pprint

import pandas as pd
from tqdm import tqdm
```

```python
evaluation_root = "/content/drive/MyDrive/w266 NLP Final Project/Evaluation/"
filename = join(evaluation_root, "evaluation_database.json")

# Read JSON into dataframe
evaluation_db_df = pd.read_json(filename)
```

### Quick EDA to see what we're working with

```python
evaluation_db_df.head()
```

```python
# See each prediction set name
evaluation_db_df.groupby("prediction_set").size()
```

```python
# Filter for specific prediction set
evaluation_db_df[
    evaluation_db_df["prediction_set"]
    == "predictions.bart_base_pt_long.triviaqa.squad.csv"
]
```

```python
# Find the overall averages for each prediction set and metric
eval_means_df = evaluation_db_df.groupby(["nickname"]).mean()

# Reset index so able to groupby and sort below to find top scoring prediction sets for each metric
eval_means_df = eval_means_df.reset_index("nickname")

eval_means_df
```

### Begin Analysis

Top predictions

```python
# Thanks JL for the code..

# Show top 5 prediction sets by avg metric scores
top_x_performers = 5

metrics = [
    "rouge1",
    "rouge2",
    "rougeL",
    "meteor",
    "bleurt",
    # 'bertscore-precision',
    # 'bertscore-recall',
    "bertscore-f1",
]


for met in metrics:
    tops = eval_means_df.sort_values(by=met, ascending=False)[["nickname", met]].head(
        top_x_performers
    )
    print(met + ":\n", tops, "\n")
```

```python
# Based on above results,
# filter for specific prediction set with consistently high scores
bart_nq_nq_df = evaluation_db_df[evaluation_db_df["nickname"] == "bart_nq_nq"]

# Specify columns to return and metric to look at
bart_nq_nq_df = bart_nq_nq_df[["target", "prediction", "bertscore-f1"]]

bart_nq_nq_df
```

```python
# Find top 5 predictions in bart nq nq for text review
bart_nq_nq_df_top = bart_nq_nq_df.sort_values(by="bertscore-f1", ascending=False)

bart_nq_nq_df_top
```

Bottom predictions

```python
# Thanks JL for the code..

# Show bottom 5 prediction sets by avg metric scores
bottom_x_performers = 5

metrics = [
    "rouge1",
    "rouge2",
    "rougeL",
    "meteor",
    "bleurt",
    # 'bertscore-precision',
    # 'bertscore-recall',
    "bertscore-f1",
]


for met in metrics:
    bottoms = eval_means_df.sort_values(by=met, ascending=True)[["nickname", met]].head(
        bottom_x_performers
    )
    print(met + ":\n", bottoms, "\n")
```

```python
# Based on above results,
# filter for specific prediction set with consistently low scores
t5_quac_nq_df = evaluation_db_df[evaluation_db_df["nickname"] == "T5_quac_nq"]

# Specify columns to return and metric to look at
t5_quac_nq_df = t5_quac_nq_df[["target", "prediction", "bertscore-f1"]]

t5_quac_nq_df
```

```python
# Find bottom 5 predictions in t5 quac triviaqa for text review
t5_quac_nq_df_bottom = t5_quac_nq_df.sort_values(by="bertscore-f1", ascending=True)

t5_quac_nq_df_bottom
```

### Metrics by Data Tested On

```python
# Set number of decimal places for scores in pandas df
pd.set_option('precision', 4)
```

```python
# Tested on NQ
nq_df = evaluation_db_df[evaluation_db_df['tested_on'] == 'nq'].groupby(['base_model', 'trained_on', 'tested_on']).mean().sort_values(by=['bertscore-f1'], ascending=False)

nq_df
```

```python
import plotly.graph_objects as go


# Reset index
data = nq_df.reset_index()
# Specify columns to return and metric to look at
data = data[["base_model", "trained_on", "bertscore-f1"]]


fig = go.Figure(data=go.Heatmap(
                   z=data['bertscore-f1'],
                   x=data['base_model'],
                   y=data['trained_on'],
                   hoverongaps = False,
                   text=data['bertscore-f1'],
                   texttemplate="%{text}",
                   colorscale='Viridis'))
fig.show()
```
